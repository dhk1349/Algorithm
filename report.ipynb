```python
import random 
import time
import math
import seaborn as sns
import pandas as pd
import numpy as np
from prob1 import *
from tqdm import tqdm
```

```python
def timer(sort_alg, lst):
    start = time.time()
    sort_alg(lst)
    end = time.time()
    return end-start
```

```python
lst=[3,2,1]
sort_alg = [bubble_sort]
for i in sort_alg:
    print(timer(i, lst))
```

## Bubble Sort

```python
x = []
threshold = []
y = []
# nlogn = []
size = 1000/1.2
for i in tqdm(range(1, 10)):
    size *= 1.2
    lst = [random.randint(1, 1000000000) for _ in range(int(size))]
    for alg in sort_alg:
        t=timer(bubble_sort, lst)
    x.append(int(size))
    y.append(t)
    if i==1: #unit length is 1000
        unit_t=t
        unit_size=1000
    threshold.append(unit_t*pow(size ,2)/pow(unit_size, 2))
    # nlogn.append(math.log(int(size), 2)* int(size))
    # print(f"Size of list: {int(size)}\nTime: {t}\n")
print(f"Time needed to sort longest list: {t}")
```

```python
df = pd.DataFrame(dict(size=x ,time=y, label="buuble"))
to_add = pd.DataFrame(dict(size=x, time =threshold, label="threshold"))
```

```python
df=pd.concat([df, to_add])
```

```python
g = sns.relplot(x="size", y="time", kind="line", hue="label", style="label", markers=True, data=df)
g.fig.autofmt_xdate()
```
